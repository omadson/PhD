```python
import json
import pandas as pd
from shapely.geometry import shape, Point

# load GeoJSON file containing the neighborhood
with open('data/limitebairro.json') as f:
    js = json.load(f)

# load CSV with stations
fields = ['id', 'lat', 'lon', 'program', 'neighborhood', 'etapa']
stations = pd.read_csv('data/bicicletar/estacoesCompletoExtended.csv', usecols=fields, index_col='id')
stations['coordinates'] = list(zip(stations.lon, stations.lat))
stations['coordinates'] = stations['coordinates'].apply(Point)


count  = dict()
stations = stations.assign(new_neighborhood=stations['neighborhood'])
# check each polygon to see if it contains the point
for feature in js['features']:
    polygon = shape(feature['geometry'])
    count[feature['properties']['NOME']] = [0,0]
    for (point, program, index) in zip(stations['coordinates'],stations['program'], stations.index):
        if polygon.contains(point):
            stations.loc[stations.index == index, 'new_neighborhood'] = feature['properties']['NOME']
            if program == 'Bicicletar':
                count[feature['properties']['NOME']][0] = count[feature['properties']['NOME']][0] + 1
            if program == 'Minibicicletar':
                count[feature['properties']['NOME']][1] = count[feature['properties']['NOME']][1] + 1

count_stations = pd.DataFrame([(a,b[0],b[1]) for (a,b) in list(count.items())], \
             columns=['neighborhood', 'bicicletar', 'minibicicletar'])

count_stations.to_csv('data/count_stations.csv', index=False)
stations.to_csv('data/bicicletar/stations_with_new_neighborhood.csv')
```

```python

count  = dict()
# check each polygon to see if it contains the point
for feature in js['features']:
    polygon = shape(feature['geometry'])
    count[feature['properties']['NOME']] = [0,0,0,0,0,0]
    for (point, program, index, stage) in zip(stations['coordinates'],stations['program'], stations.index, stations['etapa']):
        for etapa in stations['etapa'].unique():
            if etapa == stage:
                if polygon.contains(point):
                    count[feature['properties']['NOME']][etapa] = count[feature['properties']['NOME']][etapa] + 1

count_stations = pd.DataFrame([(a,b[0],b[1], b[2], b[3], b[4], b[5]) for (a,b) in list(count.items())], \
              columns=['neighborhood', 'stg_0', 'stg_1', 'stg_2', 'stg_3', 'stg_4', 'stg_5'])

count_stations.to_csv('data/count_stations_by_stg.csv' % etapa, index=False)
```
